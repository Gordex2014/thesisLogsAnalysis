```python
import numpy as np
import matplotlib.pyplot as plt


N = 2
years_18 = np.array([4, 14])
years_26 = np.array([7, 1])
years_56 = np.array([4, 0])

ind = np.arange(N)
width = 0.55

plt.figure(figsize=(4,5))

p1 = plt.bar(ind, years_18, width, yerr=0, color='#03045E')
p2 = plt.bar(ind, years_26, width, bottom=years_18, yerr=0, color='#0077B6')
p3 = plt.bar(ind, years_56, width, bottom=years_18+years_26, yerr=0, color='#00B4D8')

plt.xticks(ind, ('Grupo focal 1', 'Grupo focal 2'))
plt.yticks(np.arange(0, 16, 1))
plt.legend((p1[0], p2[0], p3[0]), ('18 años - 25 años', '26 años - 55 años', '56 años o más'), loc='center left', bbox_to_anchor=(1, 0.5))

plt.grid(color='black', linestyle='-', linewidth=1, axis='y')

plt.savefig('pregunta_1_encuesta.png', dpi=300)

plt.show()
```
